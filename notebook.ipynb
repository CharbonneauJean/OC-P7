````python
# open as pandas the heart_2022_no_nans.csv file
import pandas as pd

df = pd.read_csv('data/heart_2022_no_nans.csv')
````

````python
# drop columne 'State'
df.drop('State', axis=1, inplace=True)
````

````python
df.info()
````

````python
df.sample(10)
````

````python
# extract df.sample(10) as md with table format markdown
def extract_sample(df, n=10):
    return df.sample(n).to_markdown(index=False)

mdfile = extract_sample(df)

# write "sample_data.md" file
with open("sample_data.md", "w") as f:
    f.write(mdfile)
````

````python
# ... existing code ...

# New cell to write unique values of non-numeric columns to a markdown file
output_md_content = ""
for col in df.columns:
    if df[col].dtype == 'object': # Checks if the column is non-numeric
        output_md_content += f"## Column: {col}\n"
        output_md_content += f"Number of unique values: {df[col].nunique()}\n"
        output_md_content += "Unique values:\n"
        output_md_content += "```\n"
        for val in df[col].unique():
            output_md_content += f"{val}\n"
        output_md_content += "```\n"
        output_md_content += "---\n\n"

with open("string_unique_values.md", "w", encoding='utf-8') as f:
    f.write(output_md_content)

print("Markdown file 'string_unique_values.md' has been created.")

# ... existing code ...
````
